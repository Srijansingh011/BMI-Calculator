```python
name = input("Enter Your name")

weight =int(input("Enter your weight in Kilo Grams: "))

height =int(input("Enter your height in Centimeters: "))

BMI = (weight*1550)/(height * height)

print(BMI)

if BMI>0:
    if(BMI<18.5):
        print(name +", You are Underweight.")
    elif(18.5<=BMI<=24.9):
        print(name +", You are Normal (Healthy).")
    elif(25<=BMI<=29.9):
        print(name +", You are Overweight. You need to excercise more")
    elif(30<=BMI<=34.9):
        print(name +", You are Obese. You really need some treatment plans")
    elif(35<=BMI<39.9):
        print(name +", You are Severely Obese.")
    elif(35<=BMI<39.9):
        print(name +", You are Mordibly Obese .")
    else:
        print("Enter Valid Inputs")
```

```python
BMI Category	       BMI Range	
Underweight	           Under 18.5	
Normal (Healthy)	   18.5 – 24.9
Overweight	           25 – 29.9	
Obese	               30 - 34.9
Severely Obese         35 - 39.9
Mordibly Obese         40 and Over
```
